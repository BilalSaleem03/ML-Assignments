#Training ANN Model

```python
#Splitting data

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#import required functions and classes
import tensorflow as tf
from tensorflow.keras.models import Sequential    #to implement neutral network as sequence of layers to implement forward and backward propogation
from tensorflow.keras.layers import Dense         #to create nodes of neutal network
from tensorflow.keras.layers import Dropout       #to eliminate useless nodes while model
from tensorflow.keras.layers import LeakyReLU , ReLU   # Activation functions
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.optimizers import Adam
```

```python
#Early stoping (so i can stop training model when the val_loss become almost constant which means model is not become efficient anymore)
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=1,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
def create_car_price_model(input_dim):
    model = Sequential([
        # Process all features
        Dense(256, activation='relu', input_shape=(input_dim,)),
        # LeakyReLU(alpha=0.01),  # Activation as a separate layer
        BatchNormalization(),
        Dropout(0.4),

        # Dense(256),
        # LeakyReLU(alpha=0.01),
        # BatchNormalization(),
        # Dropout(0.3),

        Dense(128, activation='relu'),
        # LeakyReLU(alpha=0.01),
        BatchNormalization(),
        Dropout(0.3),

        # Learn patterns
        Dense(64, activation='relu'),
        # LeakyReLU(alpha=0.01),
        BatchNormalization(),
        Dropout(0.2),

        # Final computation
        Dense(32, activation='relu'),
        # LeakyReLU(alpha=0.01),
        BatchNormalization(),
        Dropout(0.1),

        # Output price
        Dense(1 , activation='linear')
    ])
    return model
```

```python

from tensorflow.keras.models import load_model

try:
  loaded_ann_model = load_model('/content/drive/MyDrive/Trained Models/ANN_electricity_Price_Prediction_model.h5')
  print("ANN model loaded successfully.")
  classifier = loaded_ann_model
except:
  print("Model is Training....")
  model = create_car_price_model(x_train.shape[1])
  model.compile(optimizer=Adam(), loss='huber', metrics=['mae', 'mse'])
  model_history = model.fit(x_train, y_train, validation_split=0.33, batch_size=32, epochs=100 , callbacks=early_stopping)
  # model_history = model.fit(x_train, y_train, validation_split=0.33, batch_size=10, epochs=100)
```

```python
#Loss Graph

plt.plot(model_history.history['loss'] , label='train_loss')
plt.plot(model_history.history['val_loss'] , label='val_loss')
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
plt.show()
```

```python
#predicting the tese set result

y_pred = model.predict(x_test)
```

```python
#graph b/w actual and prediction values of test data

plt.scatter(y_test, y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```

#XGBoost Model

```python
# XGBoost implementation
import xgboost as xgb
from sklearn.model_selection import train_test_split

# Split data
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Train XGBoost
xgb_model = xgb.XGBRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

xgb_model.fit(X_train, Y_train,
          eval_set=[(X_test, Y_test)],
          verbose=False)

# Predict and evaluate
predictions = xgb_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Evaluate the model
xgb_r2 = r2_score(Y_test, predictions)
xgb_mae = mean_absolute_error(Y_test, predictions)
xgb_mse = mean_squared_error(Y_test, predictions)

print(f"XGBoost R2 Score: {xgb_r2:.4f}")
print(f"XGBoost MAE: {xgb_mae:.4f}")
print(f"XGBoost MSE: {xgb_mse:.4f}")
```
