#Plotting with Matplotlib

* Import matplotlib.pyplot as plt.
* Create line plots using plt.plot().
* Create bar plots using plt.bar().
* Create scatter plots using plt.scatter().
* Create histograms using plt.hist().

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('drive/MyDrive/Data Sets/Matplotlib-DataSets/batter.csv')
```

```python
sample = df.head(30)
sample
```

```python
plt.plot(sample['runs'] , sample['avg'])
```

```python
plt.bar(sample['batter'] , sample['avg'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.scatter(sample['runs'] , sample['avg'])
```

```python
plt.hist(sample['avg'] ,bins=[20,25,30,35,40,45,50 ], width=1)
plt.show()
```

#Plot Customization


* Add titles, axis labels, legends using plt.title(), plt.xlabel(), plt.ylabel(), plt.legend().
* Use plt.grid() and customize line color and styles.

```python
plt.plot(sample['runs'] , sample['avg'] , color='red' ,linewidth=3 , marker='o' , label="runsbyavg")
# plt.plot(sample['runs'] , sample['strike_rate'] , color='blue' ,linewidth=3 , marker='+')
plt.title("Comparisen b/w runs and average and strike rate")
plt.xlabel("runs")
plt.ylabel("average")
plt.legend()
plt.grid()
```

# Subplots and Multiple Figures

###Use plt.subplot() to create multiple plots in one figure

```python
fig , ax = plt.subplots(nrows=2 ,  ncols=1 , sharex=True , figsize=(10,7))   #shareX is because thes ate sharing x axis

ax[0].scatter(sample['avg'],sample['strike_rate'],color='red',marker='+')
ax[1].plot(sample['avg'],sample['runs'],color='blue',marker='+')
```

###Plot multiple datasets on the same graph with different styles.

```python
batsman = pd.read_csv('drive/MyDrive/Data Sets/Matplotlib-DataSets/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'] ,  batsman['V Kohli'] , color='green' , linestyle='solid' ,linewidth=5 , marker='D' , markersize=10 , label='Virat')
plt.plot(batsman['index'] ,  batsman['RG Sharma'] , color='black' , linestyle='dashed',linewidth=3 , marker='o' , label='Rohit')


plt.title("IPL runs comparison b/w Virat and Rohit")
plt.xlabel("IPL Years")
plt.ylabel("Runs Scored")


plt.legend()
```
