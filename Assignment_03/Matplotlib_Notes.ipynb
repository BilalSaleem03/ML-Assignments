#Bar Chat

* Bivariate Analysis
* Numerical vs Categorical
* Use case - Aggregate analysis of groups

```python
children = [10 , 20 , 14 , 17]
colors = ['red' , 'blue' , 'green' , 'pink']
plt.bar(colors , children , color='black')
```

```python
#Its bydefault a vertical chart , but we can make horizontal chart

children = [10 , 20 , 14 , 17]
colors = ['red' , 'blue' , 'green' , 'pink']
plt.barh(colors , children , color='black')
```

#Pie Graph

* Univariate/Bivariate Analysis
* Categorical vs numerical
* Use case - To find contibution on a standard scale

```python
data = [10,40,22,100]
subjects = ['eng' , 'math' , 'sci' , 'is']

plt.pie(data , labels=subjects)
plt.show()
```

#SubPlots
* when we want to plot multiple graphs side by side

```python
batters = pd.read_csv('drive/MyDrive/Data Sets/Matplotlib-DataSets/batter.csv')
```

```python
#normalapproach to plot graphs
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
#different way which we also use in subplots
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
#now do subplots
fig , ax = plt.subplots(nrows=2 ,  ncols=1)
```

```python
fig , ax = plt.subplots(nrows=2 ,  ncols=1 , sharex=True , figsize=(10,4))   #shareX is because thes ate sharing x axis

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax[1].scatter(batters['avg'],batters['runs'],color='red',marker='+')
```
